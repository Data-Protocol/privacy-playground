## Introduction

Elite Cars is a new ride sharing service for the rich and famous.  Privacy and discretion is of the utmost importance. It has undergone rapid growth since launch 6 months ago and did not pay attention to the any data privacy risks. 

We shall use Elite Cars as our working example to introduce the concepts on how to apply Data Privacy. 

Data Privacy is all about managing risk. Not all data has equal risk as discussed and different data should be treated differently.  We have decided to define 4 different categories of data.

| Category | Description | 
| :------: | :---------- |
| Restrictive| Data that is highly dangerous to an individual if compromised.  For example, it can uniquely identify an individual without any ambiguity, it can be used to impersonate an individual in security checks, it can be used to impersonate ownership, or is highly sensitive if made public such as sexual orientation, healthcare history, political beliefs |
| Confidential | Data associated to an individual that can potentially identify them when combined with other data either held in the company or freely available publically.  This is data that is dangerous when used cleverly to fool other people and learn more about individuals they may not want to be known |
| Internal | Data that can be freely available inside a company and with any partners under NDA |
| Public | Data that can be freely shared outside the company with no risk of individual privacy compromise | 

Data Privacy is not a black and white instrument.  Decisions have to be made on appropriate data controls to put in place to ensure which data is collected, how visible that data is to others, and how long that data is kept before deletion.  There are three basic strategies on how to implement data privacy.

| # | Strategy | Introduction | 
| :-:| :------: | :---------- |
| 1 | Don't collect the data | The best strategy is to collect as little data as possible.  The challenge with data is that once collected and placed into an organisation, it spreads, pollutes and becomes very difficult to control.  For most existing companies, they already have data that is potentially dangerous and needs to be managed. |
| 2 | Delete data you no longer need | The next strategy is to find all data you know is a risk liability rather than an asset.  It should be deleted or reduced in its specificity (lower specificity through aggregation). |
| 3 | Control access to the data | The last strategy is to control access to data so only those that need the data can see the data, and there is visibility on who is accessing the data, when, where and why. |

These three strategies are complementary. Number 3 helps number 2 help number 1 which then helps number 2 which then helps number 3 and so on.  They should co-exist to create the wanted long term management and operational profile. For existing companies, locating, categorising and structuring data appropriately to enable the appropriate access for risk profile wanted.  Having high risk data grouped with all other data means all data needs to be managed at a high risk level. If you keep poison in your fridge, you better make sure your young children never open the door... 

In the past, the unfortunate answer for most companies was to collect as much data as possible and to keep the data forever. That has led to the privacy breaches and privacy abuse recently seen, not mentioning large operational expense and lack of performance. Data regulation is driving companies to have to have a higher quality of data management.  All future companies will implement high quality data management, since the alternative in terms of fines, reputational cost and operational overhead, will not be sustainable.
